```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the image
image = Image.open('C:\\Users\\Prakruthi P D\\OneDrive\\Desktop\\imageJ.jpg')  # Update the path if needed

# Convert the image to RGB (in case it's not)
image = image.convert('RGB')

# Split the image into R, G, B channels
r, g, b = image.split()

# Calculate histograms for each channel
r_hist = np.histogram(np.array(r).flatten(), bins=256, range=(0, 255))[0]
g_hist = np.histogram(np.array(g).flatten(), bins=256, range=(0, 255))[0]
b_hist = np.histogram(np.array(b).flatten(), bins=256, range=(0, 255))[0]

# Plot histograms
plt.plot(r_hist, '--r', label='Red')
plt.plot(g_hist, '--g', label='Green')
plt.plot(b_hist, '--b', label='Blue') 
plt.title('Histogram of Intensity Values')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
